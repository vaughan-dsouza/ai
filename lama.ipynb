```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
import torch

def chat_with_model(prompt, max_tokens=100):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_new_tokens=max_tokens)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response
```

```python
print("TinyLlama Chat (type 'exit' to stop)")
history = ""
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        break
    # Keep conversation context
    history += f"You: {user_input}\nAI:"
    response = chat_with_model(history, max_tokens=100)
    print(response.replace(history, "").strip())
    history += f" {response.replace(history, '').strip()}\n"
```

## Download the gguf model

### Subtask:
Download the TinyLlama GGUF model file.


**Reasoning**:
I need to download the specified TinyLlama GGUF model file from the Hugging Face Hub. The `hf_hub_download` function from the `huggingface_hub` library is the appropriate tool for this task.



```python
from huggingface_hub import hf_hub_download

model_path = hf_hub_download(
    repo_id="TheBloke/TinyLlama-1.1B-Chat-v1.0-GGUF",
    filename="tinyllama-1.1b-chat-v1.0.Q5_K_M.gguf"
)

print(f"Downloaded model path: {model_path}")
```

## Load the model

### Subtask:
Load the downloaded GGUF model using `llama-cpp-python`.


**Reasoning**:
Import the Llama class and instantiate it with the downloaded model path.



```python
from llama_cpp import Llama

llm = Llama(model_path=model_path)
```

## Implement chat functionality

### Subtask:
Create a function or loop to interact with the loaded model.


**Reasoning**:
Implement the chat loop to interact with the loaded model.



```python
print("TinyLlama Chat (type 'exit' to stop)")
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        break
    response = llm(user_input, max_tokens=100)
    print("AI:", response["choices"][0]["text"].strip())
```
